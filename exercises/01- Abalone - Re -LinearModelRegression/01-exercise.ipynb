```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("abalone.data",header=None)
df.columns=['Sex','Length','Diameter','Height','Whole_weight','Shucked_weight','Viscera_weight','Shell_weight','Rings']
df.info()
```

```python
df.head(5)
```

```python
# removing Missing value ?
df.replace('?', pd.NA, inplace=True)
df.dropna(inplace=True)

# Converting string record to intiger Sex feature
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])

df.head(5)
```

```python

# Overwriting the original DataFrame in selecting important features
df = df[['Height','Whole_weight','Shucked_weight','Shell_weight','Rings']]


# xy train test 
predict = "Rings"
X = np.array(df.drop([predict], axis=1))
Y = np.array(df[predict])
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,random_state=123)# seed is used 



# make model after reciving your train and test vars
model = linear_model.LinearRegression()
# fit the model that gets only Trains then calculate acoureccy that gets only tests
model.fit(x_train,y_train)
acc=model.score(x_test,y_test)

# show acc
print(acc)
```

```python

# Plot 
y_pred = model.predict(x_test)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title('Actual vs Predicted Rings')
plt.xlabel('Actual Rings')
plt.ylabel('Predicted Rings')
plt.grid()
plt.show()
```
